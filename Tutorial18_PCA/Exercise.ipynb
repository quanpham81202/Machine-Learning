```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.describe()
```

Z score chỉ ra cho chúng ta biết điểm dữ liệu cách xa bao nhiêu so với độ lệch chuẩn

Ví dụ trong trường hợp mà chúng ta có ở dữ liệu đầu tiên giá trị Cholesterol = 289 và mean của Cholesterol là 198.799, std của Cholesterol là 109.384
Công thức để tính Z = (X - mean)/std = (289 - 198.799)/109.384 = 0.82

Và nếu như Z score nằm ngoài (-3,3) điều đó có nghĩa là dữ liệu đó cần được loại bỏ bởi vì nó có khả năng gây nhiễu data trong quá trình huấn luyện

```python
df.shape
```

```python
df = df[df['Cholesterol'].apply(lambda x: x < 3*df.Cholesterol.std()+df.Cholesterol.mean())]
df
```

```python
df = df[df['MaxHR'].apply(lambda x: x < 3*df.MaxHR.std()+df.MaxHR.mean())]
df
```

```python
df = df[df['FastingBS'].apply(lambda x: x < 3*df.FastingBS.std()+df.FastingBS.mean())]
df
```

```python
df = df[df['Oldpeak'].apply(lambda x: x < 3*df.Oldpeak.std()+df.Oldpeak.mean())]
df
```

```python
df = df[df['RestingBP'] < 3*df.RestingBP.std()+df.RestingBP.mean()]
df
```

```python
df.shape
```

Sau khi áp dụng ouliner z score thì data của chúng ta còn 902 dòng và 12 cột so với trước đó là 918 dòng

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
X = df.drop(['HeartDisease'], axis = 'columns')
y = df.HeartDisease
```

SCALING USING STANDARDSCALER

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled.shape
```

```python
from sklearn.model_selection import train_test_split

X_train_scaled, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
##LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train)
lr_model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(), X_scaled, y, cv = 5)
```

```python
##SVM
svm_model = SVC()
svm_model.fit(X_train_scaled, y_train)
svm_model.score(X_test, y_test)
```

```python
cross_val_score(SVC(), X_scaled, y, cv = 5)
```

```python
##Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train_scaled, y_train)
rf_model.score(X_test, y_test)
```

```python
cross_val_score(RandomForestClassifier(), X_scaled, y, cv = 5)
```

PCA METHOD

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_train_pca, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_pca, y_train)
lr_model.score(X_test, y_test)
```

```python
cross_val_score(LogisticRegression(), X_pca, y, cv = 5)
```

```python
svm_model = SVC()
svm_model.fit(X_train_pca, y_train)
svm_model.score(X_test, y_test)
```

```python
cross_val_score(SVC(), X_pca, y, cv = 5)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train_pca, y_train)
rf_model.score(X_test, y_test)
```

```python
cross_val_score(RandomForestClassifier(), X_pca, y, cv = 5)
```
