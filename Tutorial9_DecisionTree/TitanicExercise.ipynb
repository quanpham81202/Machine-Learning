```python
import pandas as pd 
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
target = df.Survived
```

```python
inputs = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 'columns')
```

```python
inputs
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex = LabelEncoder()
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
```

```python
inputs['Sex_n']
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 0}) ##Đối với dữ liệu ít thì ta có thể sử dụng map để set kiểu dữ liệu cho chúng thay vì phải LabelEncodeer
```

```python
inputs
```

```python
x = inputs.drop(['Survived'], axis = 'columns')
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, target, test_size=0.2, random_state = 42)
```

```python
len(X_train)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```
